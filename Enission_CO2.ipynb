```python
data_clean = dataset.drop(columns=['MODELYEAR'])
```

According to my domain expert:

```python
data_clean = data_clean.drop(columns=['MAKE', 'MODEL'])
```

```python
data_clean = data_clean.fillna(data_clean.mean())
```

```python
categorical_columns = ['VEHICLECLASS', 'TRANSMISSION', 'FUELTYPE']
dummy_columns = pd.get_dummies(data_clean[categorical_columns], dtype=float)
data_clean = data_clean.drop(columns=categorical_columns)
data_clean = pd.concat([data_clean, dummy_columns], axis=1)
data_clean.shape
```

```python
X = data_clean.drop(columns=['CO2EMISSIONS'])
y = data_clean['CO2EMISSIONS']
print('Shape of X: ', X.shape)
print('Shape of y: ', y.shape)
```

```python
# from sklearn.model_selection import train_test_split

train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.2, random_state=4)
print(train_X.shape)
print(train_y.shape)
print(test_X.shape)
print(test_y.shape)
```

```python
# from sklearn.linear_model import LinearRegression

model_line_reg = LinearRegression()
model_line_reg.fit(train_X, train_y)
```

```python
coefficients = pd.DataFrame({'Feature': train_X.columns,
                             'Coefficient': model_line_reg.coef_})
coefficients
```

```python
pred_y_line_reg = model_line_reg.predict(test_X)
```

```python
# from sklearn.metrics import r2_score

r2_score(test_y, pred_y_line_reg)
```

It is so good. But can we make it better?

```python
# import statsmodels.api as sm

model_ols = sm.OLS(train_y, train_X).fit()
print(model_ols.summary())
# model_ols.fit(train_X, train_y)
```

We observe that: `ENGINESIZE` , `FUELCONSUMPTION_HWY`, their p-value is high so we can drop them:

```python
data_clean = data_clean.drop(columns=['ENGINESIZE', 'FUELCONSUMPTION_HWY'])

X = data_clean.drop(columns=['CO2EMISSIONS'])
y = data_clean['CO2EMISSIONS']

train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.2, random_state=4)
```

```python
model_line_reg = LinearRegression()
model_line_reg.fit(train_X, train_y)

pred_y = model_line_reg.predict(test_X)
r2_score(test_y, pred_y)
```

The result is a little better!
